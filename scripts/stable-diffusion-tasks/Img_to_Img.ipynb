```python
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image
import torch

# Load pipeline
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")

print("✅ Img2Img pipeline ready.")
```

```python
from IPython.display import display

input_image = Image.open("/content/a-cosy-street-i-drew-inspired-by-studio-ghibli-works.webp").convert("RGB").resize((512, 512))

prompt = "A ghibli themed street"
guidance_strength = 0.75  # 0.3 = more like input, 0.9 = more like prompt

output = pipe(prompt=prompt, image=input_image, strength=guidance_strength, num_inference_steps=50)
image = output.images[0]

display(image)
```
